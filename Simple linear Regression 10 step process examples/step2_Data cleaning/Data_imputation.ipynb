# K-Nearest Neighbors imputation 

```python
k is small:
    local estimation

k is large:
    global estimation
```

```python
to calculate similarity:
    Euclidean distance,
    Manhattan distance,
    cosine similarity
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
# Create a sample DataFrame with missing values
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, 7, 8],
                   'C': [9, 10, 11, np.nan]})
print("Original DataFrame:")
print(df)

# split the data into complete and incomplete data
complete_data = df.dropna()
print(complete_data)
incomplete_data = df[df.isnull().any(axis=1)]
print(incomplete_data)

# check if there are any missing values in the DAtaframe
if incomplete_data.empty:
    print("No missing values in the DataFrame")
else:
    # choose the k-value
    k = 3
    
    # perform KNN imputation
    imputer = KNNImputer(n_neighbors=k)
    print(imputer)
    imputed_values = imputer.fit_transform(incomplete_data)
    
    # create a DataFrame from the imputed values
    imputed_df = pd.DataFrame(imputed_values, columns=incomplete_data.columns)
    
    # replace the missing values in the original dataframe
    df[incomplete_data.columns] = imputed_df
    
    print("DataFramee after KNN imputation: ")
    print(df)
```

# Hot-Deck imputation

```python
nearest neighbors are determined based on similarity measures such as 
1. distance
2. similarity score, other relevant criteria
```

```python
df
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, 7, 8],
                   'C': [9, 10, 11, np.nan]})

print("Original DataFrame:")
print(df)


# Iterate through each row in the dataframe

for i, row in df.iterrows():
    # Check if the row has any missing values
    if row.isnull().any():
        # find the nearest neighbors for the row
        neighbors = df.iloc[i-1:i+2].dropna()
        # use the values of the nearest neighbors to fill in the missing values
        for col in df.columns:
            if pd.isnull(row[col]):
                df.at[i, col] = neighbors.sample(1)[col].values[0]

print("DataFrame after hot-deck imputation: ")
print(df)
```

```python
# Create a sample dataframe
df = pd.DataFrame({'Name': ['John', 'Alice', 'Bob', 'Mike'],
                   'Age': [30, None, 25, 35],
                   'Gender': [None, 'F', 'M', 'F'],
                   'Salary': [5000, None, None, None],
                   'City': ['New York', 'Los Angeles', 'Chicago', 'San Francisco']})

# Display the original dataframe
print("Original DataFrame: ")
print(df)

# calculate the percentage of missing data in each column
missing_percentage = df.isnull().sum()/len(df) * 100

# drop columns with more than 60% of missing data 
threshold = 60

columns_to_drop = missing_percentage[missing_percentage > threshold].index

df.drop(columns=columns_to_drop, inplace=True, axis=1)

# display the dataframe after dropping columns
print("\nDataFrame after dropping columns: ")
print(df)
```
