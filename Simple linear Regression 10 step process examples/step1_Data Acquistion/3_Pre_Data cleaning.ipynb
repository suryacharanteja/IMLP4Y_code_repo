### 1. Handling Missing data

```python
import pandas as pd

# load the dataset
data = pd.read_csv("./plants_dc.csv")

# Impute missing values with the mean
data["growth_rate"].fillna(data["growth_rate"].mean(), inplace=True)

# deletion
data["humidity"].dropna()

if __name__ == "__main__":
    print(data)
    print(data.isnull().sum())
    print(data.isna().sum())
    

```
