# 1. RFE

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Step 1: Load and Prepare the Data
data = pd.read_csv('sentiment_data.csv')  # Load the text data into a DataFrame
X = data['text']  # Extract the text samples
y = data['label']  # Extract the sentiment labels

# Step 2: Vectorize the Text Data
vectorizer = TfidfVectorizer()  # Create a TfidfVectorizer instance
X_tfidf = vectorizer.fit_transform(X)  # Convert text data into TF-IDF vectors

# Step 3: Initialize the Model
model = LogisticRegression()  # Create a Logistic Regression model

# Step 4: Initialize the RFE Selector
rfe = RFE(estimator=model, n_features_to_select=5)  # Set the number of features to select

# Step 5: Perform Recursive Feature Elimination
rfe.fit(X_tfidf, y)

# Step 6: Analyze the Results
selected_features = vectorizer.get_feature_names_out()[rfe.support_]

# Step 7: Evaluate and Validate the Model
X_selected = X_tfidf[:, rfe.support_]

# Print the selected features
print("Selected Features:")
for feature in selected_features:
    print(feature)
```
