```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy import stats
import matplotlib.pyplot as plt

# loading and setting up the data
data = pd.read_csv("./Salary_Data.csv")
y = data["Salary"]
X = data[["YearsExperience"]]

# performing linear regression
model = LinearRegression()
model.fit(X, y)

# extract the coefficients
coef =model.coef_[0]
intercept = model.intercept_
# y = 9449.962321455074*X + 25792.20019866871

# performing Statistical significance test using z-test
# formulate hypotheses
# H0 = there is no significant relationship between salary and years of experience
# H0 ==> coefficient = 0

Hypothesized_coeff = 0
# Ha = there is a significant relationship between salary and years of experience, coefficient is not zero


# calculate the standard deviation of target variable
n = len(X)
sample_target_std = np.std(y, ddof=1)  

# calculate z-score
z_score = (coef - Hypothesized_coeff) / (sample_target_std / np.sqrt(n))

# compute the p-value
p_value = stats.norm.sf(np.abs(z_score)) * 2

# comparing p-value with significance level
significance_level = 0.05

if p_value < significance_level:
    print("Reject the null hypothesis. the 'YearExperience' feature is statistically signficant.") 
else:
    print("Fail to Reject the null hypothesis. the 'YearExperience' feature is not statistically signficant.")       


# plot the relation
plt.scatter(X, y, color="black")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("scatter plot of Salary and YearExperience")
plt.show()
```
