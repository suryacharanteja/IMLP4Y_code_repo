# Univariate selection using Chi-Square test for categorical variable

```python
import pandas as pd
import scipy.stats as stats

# Create the contingency table from your dataset
data = {
    'Shipping Method': ['Standard', 'Standard', 'Express', 'Same-Day', 'Standard', 'Express', 'Express', 'Same-Day'],
    'Satisfaction Level': ['Satisfied', 'Very Satisfied', 'Satisfied', 'Neutral', 'Neutral', 'Dissatisfied', 'Dissatisfied', 'Satisfied']
}

df = pd.DataFrame(data)

# Set up hypotheses
null_hypothesis = "There is no association between shipping method and satisfaction level."
# expected frequencies = observed frequencies    
alternative_hypothesis = "There is an association between shipping method and satisfaction level."

# Create a contingency table
contingency_table = pd.crosstab(df['Shipping Method'], df['Satisfaction Level'])

# Perform the Chi-Square Test of Independence
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)



# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Test of Independence Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))
print("\nHypotheses:")
print(f"Null Hypothesis (H0): {null_hypothesis}")
print(f"Alternative Hypothesis (Ha): {alternative_hypothesis}")

# Determine the significance based on the p-value
significance_level = 0.05

if p_value < significance_level:
    print("\nReject the null hypothesis. There is a statistically significant association between shipping method and satisfaction level.")
else:
    print("\nFail to reject the null hypothesis. There is no statistically significant association between shipping method and satisfaction level.")
```

# opening the black box of Chi-Square independence test

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the observed frequencies
observed_data = {
    'Shipping Method': ['Standard', 'Standard', 'Express', 'Same-Day', 'Standard', 'Express', 'Express', 'Same-Day'],
    'Satisfaction Level': ['Satisfied', 'Very Satisfied', 'Satisfied', 'Neutral', 'Neutral', 'Dissatisfied', 'Dissatisfied', 'Satisfied']
}
df_observed = pd.DataFrame(observed_data)

# Define the null hypothesis and alternative hypothesis
# Null hypothesis: There is no association between the two variables
# Alternative hypothesis: There is an association between the two variables
null_hypothesis = "There is no association between the shipping method and satisfaction level."
alternative_hypothesis = "There is an association between the shipping method and satisfaction level."

# Calculate the observed chi-square statistic
observed_cross_tab = pd.crosstab(df_observed['Shipping Method'], df_observed['Satisfaction Level'])
chi2, p_value, _, _ = stats.chi2_contingency(observed_cross_tab)

# Simulate the sampling distribution
n_simulations = 1000
chi2_simulations = []
for _ in range(n_simulations):
    # Generate a randomized dataset based on the null hypothesis
    randomized_cross_tab = pd.crosstab(
        np.random.permutation(df_observed['Shipping Method']),
        np.random.permutation(df_observed['Satisfaction Level'])
    )
    print(randomized_cross_tab)
    chi2_simulated, _, _, _ = stats.chi2_contingency(randomized_cross_tab)
    print(chi2_simulated)
    chi2_simulations.append(chi2_simulated)

# Plot the sampling distribution
plt.hist(chi2_simulations, bins=20, edgecolor='black')
plt.axvline(x=chi2, color='red', linestyle='--', label='Observed Chi-Square')
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Frequency')
plt.title('Sampling Distribution of Chi-Square Statistic')
plt.legend()

# Calculate the p-value
p_value_observed = 0.37528525266160834

# Display the results
print("Null Hypothesis: " + null_hypothesis)
print("Alternative Hypothesis: " + alternative_hypothesis)
print("Observed Chi-Square Statistic:", chi2)
print("Level of Significance: 0.05")
print("observed_p-value:", p_value_observed)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the observed data
data = {
    'Shipping Method': ['Standard', 'Standard', 'Express', 'Same-Day', 'Standard', 'Express', 'Express', 'Same-Day'],
    'Satisfaction Level': ['Satisfied', 'Very Satisfied', 'Satisfied', 'Neutral', 'Neutral', 'Dissatisfied', 'Dissatisfied', 'Satisfied']
}

df = pd.DataFrame(data)

# Define the observed contingency table
observed_table = pd.crosstab(df['Shipping Method'], df['Satisfaction Level'])

# Calculate the observed chi-square statistic
chi2, p_value, dof, expected_table = stats.chi2_contingency(observed_table)

# Simulate the sampling distribution
n_simulations = 10000
chi2_simulated = []
for _ in range(n_simulations):
    # Shuffle the Satisfaction Level labels while keeping the Shipping Method labels fixed
    shuffled_satisfaction = np.random.permutation(df['Satisfaction Level'])
    shuffled_table = pd.crosstab(df['Shipping Method'], shuffled_satisfaction)
    chi2_simulated.append(stats.chisquare(shuffled_table).statistic)

# Plot the sampling distribution
plt.hist(chi2_simulated, bins=50, alpha=0.5, edgecolor='black', density=True, label='Sampling Distribution')
plt.axvline(x=chi2, color='red', linestyle='--', label='Observed Chi-Square')
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Density')
plt.title('Sampling Distribution of Chi-Square Statistic')
plt.legend()

# Plot the chi-square distribution with degrees of freedom
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df=dof)
plt.plot(x, y, 'r-', label=f'Chi-Square Distribution (dof={dof})')

# Mark the observed chi-square value and level of significance
plt.annotate(f'Observed Chi-Square = {chi2:.2f}\np-value = {p_value:.4f}', xy=(chi2, 0.02),
             xytext=(chi2+2, 0.05), arrowprops=dict(arrowstyle='->'), fontsize=10)

plt.legend()
plt.show()
```
