### 1. k-means clustering

```python
import pandas as pd
from sklearn.cluster import KMeans


# Load the data
data = pd.read_csv("./Soccer_players_data1.csv")


# select the features for clustering
X = data[['Height(cm)', 'Weight(kg)']]

# perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)


# Get cluster assignments and centroid positions
cluster_labels = kmeans.labels_
centroid_positions = kmeans.cluster_centers_

# Analyze and visualize the clusters
df = pd.DataFrame(X, columns=['Height(cm)', 'Weight(kg)'])
df['cluster'] = cluster_labels

# visualizing the clusters
import matplotlib.pyplot as plt



if __name__ == "__main__":
    print(df.head())
    
    # plotting clusters
    fig, ax = plt.subplots()
    ax.scatter(X["Weight(kg)"], X['Height(cm)'], c=cluster_labels)
    ax.scatter(centroid_positions[:, 0], centroid_positions[:, 1], c='red', marker="X")
    ax.set_xlabel('Height(cm)')
    ax.set_ylabel("Weight(kg)")
    ax.set_title("K-Means Clustering")
    plt.show()
    
```

### 2. Hierarchical Clustering

###### 2.1 Agglomerative Clustering( Bottom-up)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage


X1 = np.array([[1,1], [3,2], [9,1], [3,7], [7,2], [9,7], [4,8], [8,3],[1,4]])

# create a subplot
fig, ax = plt.subplots()
ax.scatter(X1[:, 0], X1[:, 1], c='r')

# create a numbered labels for each point
for i in range(X1.shape[0]):
    ax.annotate(str(i), xy=(X1[i, 0], X1[i, 1]), xytext=(3, 3), textcoords='offset points')
    
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('Scatter plot of the data')
ax.set_xlim([0, 10])
ax.set_xticks(range(10))
ax.set_yticks(range(10))
plt.show()

# min(single) linkage method or approach 
min_linkage = linkage(X1, method='single', metric='euclidean')

# max(complete) linkage method or approach
max_linkage = linkage(X1, method='complete', metric='euclidean')

# centroid linkage
cent_linkage = linkage(X1, method='centroid', metric='euclidean')

# average linkage
avg_linkage = linkage(X1, method='average', metric='euclidean')

# ward's linkage
ward_linkage = linkage(X1, method='ward', metric='euclidean')

# create dendograms
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1), dendrogram(min_linkage), plt.title('Single or min linkage')

plt.subplot(2, 2, 2), dendrogram(max_linkage), plt.title('Complete or max linkage')

plt.subplot(2, 2, 3), dendrogram(cent_linkage), plt.title('Centroid linkage')

plt.subplot(2, 2, 4), dendrogram(avg_linkage), plt.title('Average linkage')

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1), dendrogram(ward_linkage), plt.title('ward linkage')


if __name__ == "__main__":
    print(X1)
    print(X1.ndim)
    print(X1.size)
    print(X1.shape)
    
    
```

```python
from sklearn.cluster import AgglomerativeClustering

ward_link_Agg = AgglomerativeClustering(n_clusters=2, linkage='ward')
ward_link_Agg.fit(X1)
print(ward_link_Agg.labels_)

# visualizing the two clusters
fig, ax = plt.subplots(figsize=(6, 6))

scatter = ax.scatter(X1[:, 0], X1[:, 1], c=ward_link_Agg.labels_, cmap='rainbow')

legend = ax.legend(*scatter.legend_elements(), title="Clusters", bbox_to_anchor=(1, 1))

ax.add_artist(legend)

plt.title("Scatter plot of clusters")

plt.show()

if __name__ == "__main":
    print(type(ward_link_Agg))
```
