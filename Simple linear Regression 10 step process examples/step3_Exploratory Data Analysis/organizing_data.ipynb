```python
FE_bins = []
for fc in dfv["Average fuel (l/100km)"]:
    if fc < 99:
        FE_bins.append('lowFuelConsumed')
    elif fc < 131:
        FE_bins.append('averageFuelConsumed')
    else:
        FE_bins.append('HightFuelConsumed')

bins = pd.DataFrame(FE_bins)
bins
```

```python
# example of categorizing data into bins or categories

ages = [22, 34, 45, 56, 67, 78, 89, 90, 95, 100]

age_bins = []

for age in ages:
    if age < 30:
        age_bins.append('20s')
    elif age < 40:
        age_bins.append('30s')
    elif age < 50:
        age_bins.append('40s')
    else:
        age_bins.append('50s+')
print("Age bins:", age_bins)
```

```python
# categorizing the data using 'cut()' method
df = pd.DataFrame({'Age': [25, 30, 45, 35, 28], 'Income': [50000, 60000, 55000, 70000, 65000]})

# define binds and labels for categorization
bins = [20, 30, 40, 50]
labels = ['20-29', '30-39', '40-49']

# add a new column 'Age Group' to DataFrame based on the 'Age' column values
df["Age Group"] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
bins = [67, 99, 130, 160]
lables = ["70-98", '99-130', '131-150']
dfv["FE_group"] = pd.cut(dfv["Average fuel (l/100km)"], bins=bins, labels=lables)
```

```python
dfv["Idling time"].describe()

bins = [0.15, 0.3, 0.4, 0.5]
labels = ["verygood", 'average', 'bad']

dfv["Idling time_groups"] = pd.cut(dfv["Idling time"], bins=bins, labels=labels)
dfv
```

```python
# categorize data using "qcut()"

# let us create a simple DataFrame
df1 = pd.DataFrame({'Income': [50000, 60000, 55000, 70000, 65000]})

# define quantiles and leabels for categorization
quantiles = [0, 0.25, 0.5, 0.75, 1]
labels = ['Q1', 'Q2', 'Q3', 'Q4']

# Add a new column "Income quantile" to DataFrame based on the "income " column
df1["Income quantile"] = pd.qcut(df["Income"], q=quantiles, labels=labels)
print(df1)
```

# 3. Data Tabulation

```python
dfv["Brake_count_quant"].unique()
# frequency table
Brake_count_freq = dfv["Brake_count_quant"].value_counts()
print(Brake_count_freq)
# contingency table
Fuel_Idling_tab = pd.crosstab(dfv["FE_group"], dfv["Idling time_groups"])
print(Fuel_Idling_tab)
```

```python
# let us create a frequency table using pandas
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'], 'Age': [25, 30, 22, 35, 28, 29, 17, 31], 'City': ['Bengaluru', 'Moscow', 'Delhi', 'Vizag', 'Chennai', 'Mumbai', 'Delhi', 'Sri nagar']}

dft = pd.DataFrame(data)

# create a frequency table of Gender
gender_freq = dft["Gender"].value_counts()
print("Frequncy table of Gender:\n", gender_freq)

# create a contindency table of Gender vs City
gender_city_ct = pd.crosstab(dft['Gender'], dft['City'])
print("Contingency table of Gender vs. City:\n", gender_city_ct)
```

### 3.1 Frequency Distribution

```python
# frequency distribution of a single column

# create a sample dataframe
dffd = pd.DataFrame({'Age': [25, 30, 22, 35, 28, 30, 35, 30, 22, 25, 22, 28]})

# calculate the frequency distribution of 'age' column
freq_dist = dffd['Age'].value_counts()

print(freq_dist)
```

```python
# frequency distribution of multiple columns

# create dataframe
dffdm = pd.DataFrame({'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'], 'Age': [25, 30, 22, 35, 28]})

# calculate frequency distribution of 'Gender' and 'Age' columns
freq_dist_gender = dffdm['Gender'].value_counts()
freq_dist_age = dffdm['Age'].value_counts()

print(freq_dist_gender)
print(freq_dist_age)
```

### 3.3 grouped relative frequency distribution

```python
import numpy as np

dfgrfd = pd.DataFrame({'Age': [25, 30, 22, 35, 28, 30, 35, 30, 22, 25, 22, 28]})

# define the bin edges fro grouping the data
bin_edges = [20, 25, 30, 35, 40]

# group the data into bins and calculate the relative frequencies
age_bins = pd.cut(dfgrfd['Age'], bins=bin_edges)
grp_rel_freq_dist = age_bins.value_counts(normalize=True, sort=False)

print(grp_rel_freq_dist)
```

### 3.4 stem and lead plot

```python
dfsl = pd.DataFrame({'Scores': [68, 72, 83, 75, 90, 85, 77, 78, 89, 92,
                               76, 81, 84, 87, 88, 95, 96, 99, 100, 72,
                               68, 76, 82, 79, 74, 87, 91, 93, 97, 98]})
# extract the tens digit (stem) and unit digit (leaf) of the scores
dfsl['stem'] = dfsl['Scores'] // 10
dfsl['Leaf'] = dfsl['Scores'] % 10

# create a "stem and leaf" plot using pandas 'group by' and 'agg' functions
stem_and_leaf_plot = dfsl.groupby('stem')['Leaf'].agg(list)

print(stem_and_leaf_plot)
```
